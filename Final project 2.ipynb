```python
import pandas as pd
import numpy as np
import matplotlib as pltm
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectFromModel
np.random.seed(0)
```

```python
df = pd.read_csv('Survey-results-2018.csv')
df.head(5)
```

```python
df.info()
```

```python
df_features = df.drop(['E', 'I', 'A', 'G', 'BAI2'], axis=1)
print('The shape of our features is:', df_features.shape)
```

```python
df_features.head()
```

```python
df_features.describe()
```

```python
features_oneHot = pd.get_dummies(df_features)
```

```python
features_oneHot.shape
```

```python
# Labels are the values we want to predict

labels = np.array(features_oneHot['BAI3'])
```

```python
# Remove the labels from the features
# axis 1 refers to the columns

features= features_oneHot.drop('BAI3', axis = 1)
features.shape
```

```python
# Convert to numpy array

features = np.array(features)
```

```python
# Split the data into training and testing sets

train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.2, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
random_forest = RandomForestRegressor(n_estimators = 10, random_state = 42)
```

```python
# Train the model on training data

random_forest.fit(train_features, train_labels);
```

```python
# Use the forest's predict method on the test data

y_pred = random_forest.predict(test_features)
```

```python
# Use the forest's predict method on the test data

y_pred = random_forest.predict(test_features)
```

```python
# Calculate the absolute errors

absolute_errors = abs(y_pred - test_labels)
```

```python
# Print out the mean absolute error (mae)

print('Mean Absolute Error:', round(np.mean(absolute_errors), 2), 'degrees.')
```

```python
# mean absolute percentage error (MAPE)

mape = 100 * (absolute_errors / test_labels)
```

```python
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
importances = random_forest.feature_importances_

std = np.std([tree.feature_importances_ for tree in random_forest.estimators_],axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(train_features.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```
